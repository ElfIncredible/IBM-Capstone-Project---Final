# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect("m4_survey_data.sqlite")

cursor = conn.cursor()
table_name = "master"
cursor.execute(f"SELECT ConvertedComp FROM {table_name}")
data = cursor.fetchall()
conn.close()
converted_comp = [row[0] for row in data if row[0] is not None]

plt.hist(converted_comp, bins=20, edgecolor='k')
plt.xlabel("Converted Compensation")
plt.ylabel("Frequency")
plt.title("Histogram of Converted Compensation")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

table_name = "master"
cursor.execute(f"SELECT Age FROM {table_name}")
data = cursor.fetchall()
conn.close()

age_data = [row[0] for row in data if row[0] is not None]

plt.boxplot(age_data, vert=True)  
plt.ylabel("Age")  
plt.title("Box Plot of Age")
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

table_name = "Master"
cursor.execute(f"SELECT Age, WorkWeekHrs FROM {table_name} WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL")
data = cursor.fetchall()
conn.close()

age_data = [row[0] for row in data]
workweekhrs_data = [row[1] for row in data]

plt.scatter(workweekhrs_data, age_data, alpha=0.5)  
plt.ylabel("Age")  
plt.xlabel("WorkWeekHrs")  
plt.title("Scatter Plot of WorkWeekHrs vs. Age")  
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

table_name = "Master"

cursor.execute(f"SELECT Age, WorkWeekHrs, CodeRevHrs FROM {table_name} WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL")
data = cursor.fetchall()

conn.close()

age_data = [row[0] for row in data]
workweekhrs_data = [row[1] for row in data]
coderevhrs_data = [row[2] for row in data]

plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
scatter = plt.scatter(workweekhrs_data, coderevhrs_data, c=age_data, cmap='viridis', alpha=0.7, s=age_data, edgecolors='k')

plt.colorbar(scatter, label='Age')

plt.xlabel("WorkWeekHrs")
plt.ylabel("CodeRevHrs")
plt.title("Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble Size = Age)")
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect("m4_survey_data.sqlite")

cursor = conn.cursor()

table_name = "DatabaseDesireNextYear"

cursor.execute(f"""
    SELECT DatabaseDesireNextYear, COUNT(*) AS Count
    FROM {table_name}
    WHERE DatabaseDesireNextYear IS NOT NULL
    GROUP BY DatabaseDesireNextYear
    ORDER BY Count DESC
    LIMIT 5
""")

data = cursor.fetchall()

conn.close()

database_names = [row[0] for row in data]
database_counts = [row[1] for row in data]

plt.figure(figsize=(8, 8))  
plt.pie(database_counts, labels=database_names, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Databases Respondents Wish to Learn Next Year")
plt.axis('equal')  

plt.legend(labels=database_names, loc="best")

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import sqlite3
import matplotlib.pyplot as plt
import numpy as np

conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

table_name = "master"

cursor.execute(f"""
    SELECT Age,
           (SELECT AVG(WorkWeekHrs)
            FROM {table_name}
            WHERE Age >= 30 AND Age <= 35
            GROUP BY Age
            ORDER BY Age
            LIMIT 1) AS MedianWorkWeekHrs,
           (SELECT AVG(CodeRevHrs)
            FROM {table_name}
            WHERE Age >= 30 AND Age <= 35
            GROUP BY Age
            ORDER BY Age
            LIMIT 1) AS MedianCodeRevHrs
    FROM {table_name}
    WHERE Age >= 30 AND Age <= 35
    GROUP BY Age
    ORDER BY Age
""")

data = cursor.fetchall()
conn.close()

age_data = [row[0] for row in data]
median_workweekhrs = [row[1] for row in data]
median_coderevhrs = [row[2] for row in data]

plt.figure(figsize=(10, 6))  

plt.bar(age_data, median_workweekhrs, label='Median WorkWeekHrs', alpha=0.7)
plt.bar(age_data, median_coderevhrs, label='Median CodeRevHrs', alpha=0.7, bottom=median_workweekhrs)

plt.xlabel("Age")
plt.ylabel("Hours")
plt.title("Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35")
plt.legend()
plt.grid(True)

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
import sqlite3
import matplotlib.pyplot as plt
import numpy as np

conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

table_name = "master"

cursor.execute(f"""
    SELECT Age,
           (SELECT AVG(ConvertedComp)
            FROM {table_name}
            WHERE Age >= 45 AND Age <= 60
            GROUP BY Age
            ORDER BY Age
            LIMIT 1) AS MedianConvertedComp
    FROM {table_name}
    WHERE Age >= 45 AND Age <= 60
    GROUP BY Age
    ORDER BY Age
""")

data = cursor.fetchall()
conn.close()

age_data = [row[0] for row in data]
median_convertedcomp = [row[1] for row in data]


plt.figure(figsize=(10, 6))  

plt.plot(age_data, median_convertedcomp, marker='o', linestyle='-')

plt.xlabel("Age")
plt.ylabel("Median ConvertedComp")
plt.title("Median ConvertedComp for Ages 45-60")
plt.grid(True)

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect("m4_survey_data.sqlite")
cursor = conn.cursor()

table_name = "master"

cursor.execute(f"""
    SELECT MainBranch, COUNT(*) AS Count
    FROM {table_name}
    WHERE MainBranch IS NOT NULL
    GROUP BY MainBranch
    ORDER BY Count DESC
""")

data = cursor.fetchall()
conn.close()

main_branches = [row[0] for row in data]
branch_counts = [row[1] for row in data]

plt.figure(figsize=(10, 6))  

plt.barh(main_branches, branch_counts, color='skyblue')
plt.xlabel("Count")
plt.ylabel("MainBranch")
plt.title("Horizontal Bar Chart of MainBranch")

plt.gca().invert_yaxis()  # Invert y-axis to display the most common branch at the top

plt.show()
```
